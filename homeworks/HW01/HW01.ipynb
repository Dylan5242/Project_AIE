1. Подключает библиотеку

```python
import numpy as np
```

2. Пример создания массива с помощью np.array

```python
arr = np.array([1, 2, 3, 4, 5]) #создаем массив из списка. Аrray сама базовая операция NumPy для создания массива
print(arr, "\n")

arr = np.zeros((3, 3)) # Массив из нулей (размер 3x3)
print(arr, "\n")

arr_two = np.ones((5, 5)) #массив из едениц (5*5)
print(arr_two, "\n")

arr = np.arange(10) #создает массив c последовательностью чисел от 0 до 9 (как for-ом пройтись)
print(arr, "\n")
```

3. Демонстрация атрибутов

```python
print(arr.shape)#выдает форму массива (длина каждой вложенной секции массивов)
print(arr_two.shape, "\n")

print(arr.dtype, "\n")#тип элементов массива
```

4. Пример индексации и срезов

```python
print(arr[1]) #элемент под индексом 1 (тоесть 2й в массиве по порядку)

print(arr_two[1, 3]) #строка с индексом 1 и столбец с индексом 3

idx = np.array([0, 2, 3]) #массив со значениями индексов
print(arr[idx], "\n") #вывод значений по индексам из массива

print(arr[2:8:2]) #со 2 по индексу по 8 элемент не включительно, каждый 2 элемент
arr_two = np.arange(100).reshape(10,10)
print(arr_two[:, 1]) #все строки, 2 столбец (1 по ииндексу)
```

5. Векторные операции и базовая статика

```python
arr = np.array([20, 30, 40, 50])
arr_two = np.arange(4) # [0 1 2 3]

print(arr + arr_two)   # сложение
print(arr - arr_two)   # вычитание
print(arr * arr_two)   # умножение
print(arr / arr_two)   # деление (деление на 0 дает inf, бесконечность)
print(arr ** arr_two)  # поэлементное возведение в степень
print(arr % arr_two, "\n")   # остаток от деления


print(arr.mean()) #Среднее значение a
print(arr_two.min()) #Минимум в arr_two
print(arr_two.max()) #Максимум в arr_two
print(arr.sum()) #Сумма элементов arr
```
